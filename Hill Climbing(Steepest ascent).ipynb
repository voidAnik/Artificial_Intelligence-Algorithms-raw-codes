```
class SteepestAscent:

  def init(self):
    state = [2, 1, 5, 0, 8, 4, 10, 0, 20, 10]
    cost = self.calculate_cost(state)
    print("INITIAL STATE = ", state)
    return state, cost

  def calculate_cost(self, state):
    cost = 0
    for i in range(len(state)-1):
      for j in range(i+1, len(state)):
        if (state[i] > state[j]):
          cost += 1
    #print("calculated state = ", state)
    #print("calculated cost = ", cost)
    return cost


  def generate_state(self, current_state, current_state_cost):
    print("current state",current_state)
    print("current state cost",current_state_cost)

    selected_state = []
    selected_cost = float('inf')
    
    for i in range(len(current_state)):
      for j in range(i+1, len(current_state)):
        new_state = list(current_state)
        new_state[i], new_state[j] = new_state[j], new_state[i]
        temp_cost = self.calculate_cost(new_state)

        if (selected_cost > temp_cost):
          selected_cost = temp_cost
          selected_state = list(new_state)

    if (selected_cost < current_state_cost):
      print("\nselected state = ", selected_state)
      print("selected cost = ", selected_cost)
      return selected_state, selected_cost
    else:
      return current_state, None


  def goal_test(self, state):
    if self.calculate_cost(state) == 0:
      return True
    else:
      return False


############ MAIN #############

if __name__ == '__main__':
  HC_Model = SteepestAscent()
  state, cost = HC_Model.init()
  iteration = 0

  while HC_Model.goal_test(state) is False:
    iteration += 1
    print("\n***Iteration-0"+str(iteration))
    state, cost = HC_Model.generate_state(state, cost)
    if cost is None:
      print("\nFINAL STATE = ",state)
      break
  if cost is not None:
    print("\nFINAL STATE = ",state)
```
