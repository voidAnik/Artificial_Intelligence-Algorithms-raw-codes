```
node_name = {0:'S', 1:'A', 2:'B', 3:'C', 4:'D'}
```

```
adjacent_nodes = [[-1, 1, 4, -1, -1], [-1, -1, 2, 5, 12], [-1, -1, -1, 2, -1], [-1, -1, -1, -1, 3], [-1, -1, -1, -1, -1]]
heuristics = [7, 6, 2, 1, 0] 
```

```
class Node:
  def __init__(self, node_no, prev_node, actual_cost, total_cost):
    self.node_no = node_no
    self.prev_node = prev_node
    self.actual_cost = actual_cost
    self.total_cost = total_cost
```

```
from queue import PriorityQueue
from dataclasses import dataclass, field
from typing import Any

# for object compare error of PriorityQueue
@dataclass(order=True)
class PrioritizedItem:
    priority: int
    item: Any=field(compare=False)

# Initial node push   
minQ = PriorityQueue()
initial_node = Node(0, None, 0, 0 + heuristics[0])
minQ.put(PrioritizedItem(initial_node.total_cost, initial_node))

while not minQ.empty():
  selected_node = minQ.get().item
  #print("selected node=",selected_node.node_no)
  if selected_node.node_no == 4:
    path = []
    print("Optimal cost =",selected_node.total_cost)
    while selected_node != None:
      path.append(node_name[selected_node.node_no])
      selected_node = selected_node.prev_node
    path.reverse()
    print("Optimal path = ", end="")
    for p in path:
      print("->", p, end="")
    break

  for i, n_no in enumerate(adjacent_nodes[selected_node.node_no]):
    #print("matrix index=",n_no)
    if n_no != -1:
      #print("adjacent",i)
      temp_actual_cost = selected_node.actual_cost + adjacent_nodes[selected_node.node_no][i]
      temp_total_cost = temp_actual_cost + heuristics[i]
      new_node = Node(i, selected_node , temp_actual_cost, temp_total_cost)
      #print("(",temp_node_no , temp_actual_cost, temp_total_cost,")")
      minQ.put(PrioritizedItem(new_node.total_cost, new_node))
  
```
