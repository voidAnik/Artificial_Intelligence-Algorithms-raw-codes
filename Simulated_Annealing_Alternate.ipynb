```
import math
import random
import copy
random.seed(0)
class SimulatedAnnealing:

  def init(self):
    state = [2, 1, 5, 0, 8, 4, 10, 0, 20, 10]
    #state = [2, 1, 3, 0, 4, 8, 10, 20, 10, 30]
    #state = random.sample(range(0, 30), 10)
    cost = self.calculate_cost(state)
    print("INITIAL STATE = ", state)
    return state, cost

  def calculate_cost(self, state):
    cost = 0
    for i in range(len(state)):
      for j in range(i+1, len(state)):
        if (state[i] > state[j]):
          cost += 1
    #print("calculated state = ", state)
    #print("calculated cost = ", cost)
    return cost


  def move_or_not(self, delta_e):
    #print("In move")
    e_delta = math.exp(delta_e)
    rand = random.uniform(0,1)
    #print("e_delta = ", e_delta, "random = ", rand)
    if 0 < rand < e_delta:
      print(e_delta, rand,"move True")
      return True
    else:
      print(e_delta, rand,"move False")
      return False


  def generate_state(self, current_state, current_state_cost):
    #print("current state",current_state)
    #print("current state cost",current_state_cost)

    generated_state = []
    delta_e = 0
    for i in range(len(current_state)):
      for j in range(i+1, len(current_state)):
        #print("it")
        generated_state = copy.deepcopy(current_state)
        generated_state[i], generated_state[j] = generated_state[j], generated_state[i]
        generated_cost = self.calculate_cost(generated_state)
        #print(generated_state, generated_cost)

        if (generated_cost < current_state_cost):
          current_state = list(generated_state)
          current_state_cost = generated_cost
        elif (generated_cost == current_state_cost):
          #print("equal")
          delta_e = -0.5
          if self.move_or_not(delta_e):
            current_state = list(generated_state)
            current_state_cost = generated_cost
        else:
          #print("greater")
          delta_e = current_state_cost - generated_cost
          if self.move_or_not(delta_e):
            current_state = list(generated_state)
            current_state_cost = generated_cost
    return current_state, current_state_cost


  def goal_test(self, state):
    if self.calculate_cost(state) == 0:
      return True
    else:
      return False


############ MAIN #############

if __name__ == '__main__':
  HC_Model = SimulatedAnnealing()
  state, cost = HC_Model.init()
  iteration = 0

  while HC_Model.goal_test(state) is False:
  #for i in range(10):
    iteration += 1
    print("\n***Iteration-"+str(iteration))
    state, cost = HC_Model.generate_state(state, cost)
    print("\nselected state = ", state)
    print("selected cost = ", cost)
    if cost is None:
      print("\nFINAL STATE = ",state)
      break
  if cost is not None:
    print("\nFINAL STATE = ",state)
```
